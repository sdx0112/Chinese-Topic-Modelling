```python
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained("THUDM/chatglm2-6b", trust_remote_code=True)
```

```python
model = AutoModel.from_pretrained("THUDM/chatglm2-6b", trust_remote_code=True, device='cuda')
```

```python
model = model.eval()
```

```python
import pandas as pd
```

```python
df_test = pd.read_csv('/content/labelled_test.csv', encoding = 'GBK')
```

```python
df_test.head()
```

```python
df_sample = pd.read_csv('/content/labelled_sample.csv', encoding = 'GBK')
```

```python
df_sample.head()
```

## Zero Shot

```python
prompt = """你需要给一段文字打上话题标签。我会给你话题列表和文字内容。话题必须从给定的话题列表里选出，而且只能给一个话题。
话题列表是：基建, 房地产, 制造业, 科技, 零售消费, 绿色发展, 医药健康, 教育, 交通运输, 出口贸易, 农业农村, 文化旅游, 就业, 民生扶贫, 能源, 物资储备, 宏观, 区域发展, 企业发展, 新冠疫情, 地方财政, 税务, 国际关系, 金融
你的回答的格式必须是：answer-话题,例如：answer-科技
请给出这段文字的话题。你的回答仅包括话题。你给出的话题必须是话题列表里面的一个，而且只能给一个话题。
文字内容是：{content}
"""
```

```python
tmp = prompt.format(content=df_test['CONTENT'][10])
```

```python
tmp
```

```python
print(tmp)
```

```python
response, history = model.chat(tokenizer, tmp, history=[], temperature=0.001)
print(response)
```

```python
def get_topic(x):
  response, history = model.chat(tokenizer, x, history=[], temperature=0.001)
  return response.split('-')[-1].split('：')[-1]
```

```python
df_test['PRED_ZERO'] = df_test.apply({lambda row: get_topic(prompt.format(content=row['CONTENT']))}, axis = 1)
```

```python
df_test['PRED_ZERO'].value_counts()
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(df_test['TOPIC'], df_test['PRED_ZERO'])
```

## Few Shot

```python
prefix_prompt = """你需要给一段文字打上话题标签。我会给你话题列表和文字内容。话题必须从给定的话题列表里选出，而且只能给一个话题。
话题列表是：基建, 房地产, 制造业, 科技, 零售消费, 绿色发展, 医药健康, 教育, 交通运输, 出口贸易, 农业农村, 文化旅游, 就业, 民生扶贫, 能源, 物资储备, 宏观, 区域发展, 企业发展, 新冠疫情, 地方财政, 税务, 国际关系, 金融
你的回答的格式必须是：answer-话题,例如：answer-科技
请给出这段文字的话题。你的回答仅包括话题。你给出的话题必须是话题列表里面的一个，而且只能给一个话题。
"""
```

```python
suffix_prompt = """文字内容是：{content}"""
```

```python
examples = []
for index, row in df_sample.iterrows():
  examples.append("""文字内容：{content}\nanswer-{topic}""".format(content = row['CONTENT'], topic = row['TOPIC']))
```

```python
example_prompt = '\n'.join(examples)
```

```python
print(example_prompt)
```

```python
few_prompt = '\n'.join([prefix_prompt, example_prompt, suffix_prompt])
```

```python
response, history = model.chat(tokenizer, few_prompt.format(content = df_test['CONTENT'][10]), history=[])
```

```python
print(response)
```

```python
df_test['PRED_FEW'] = df_test.apply({lambda row: get_topic(few_prompt.format(content=row['CONTENT']))}, axis = 1)
```

```python
df_test['PRED_FEW'].value_counts()
```

```python
accuracy_score(df_test['TOPIC'], df_test['PRED_FEW'])
```
